```python
# Exercise 2 task 4
def check_variance_est(power):
    n = 1000000
    mu = 10**power
    s = np.random.normal(mu, 1, n)
    var = 1/float(n) * np.sum(s**2) - (1/float(n) * np.sum(s))**2
    print("variance mu = 10^{}   {:.6f}".format(power, var))

check_variance_est(0)
check_variance_est(3)
check_variance_est(6)
check_variance_est(7)
check_variance_est(8)
check_variance_est(9)
check_variance_est(10)
check_variance_est(11)
```

```python
# task 3.(a)
x1 = np.array([24,3,2,31])
x2 = np.array([27,20,26,21])
x3 = np.array([30,21,27,5])
x4 = np.array([26,28,25,14])
def Norm_1(x):
    len = np.sum(x)
    return len

def Norm_2(x):
    len = (np.sum(x**2))**(0.5)
    return len

def Norm_8(x):
    len = (np.sum(x**8))**(0.125)
    return len

def Norm_inf(x):
    len = np.max(x)
    return len

print("Norm_1(x1) {:.2f}".format(Norm_1(x1)))
print("Norm_2(x1) {:.2f}".format(Norm_2(x1)))
print("Norm_8(x1) {:.2f}".format(Norm_8(x1)))
print("Norm_inf(x1) {:.2f}".format(Norm_inf(x1)))

print("Norms x2   1: {:.2f}  2: {:.2f}  8: {:.2f}  inf: {:.2f}".format(Norm_1(x2),Norm_2(x2),Norm_8(x2),Norm_inf(x2)))
print("Norms x3   1: {:.2f}  2: {:.2f}  8: {:.2f}  inf: {:.2f}".format(Norm_1(x3),Norm_2(x3),Norm_8(x3),Norm_inf(x3)))
print("Norms x4   1: {:.2f}  2: {:.2f}  8: {:.2f}  inf: {:.2f}".format(Norm_1(x4),Norm_2(x4),Norm_8(x4),Norm_inf(x4)))
```
